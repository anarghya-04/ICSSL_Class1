```python
scores = [65,78,30,85]
avg_scores = sum(scores)/len(scores) #Calculate the average
max_score = max(scores) #maximum value
min_score = min(scores) #minimun value
print(f"The average score of this class is {avg_scores}, the highest score is {max_score}, and the lowest score is {min(scores)}.")
```

Functions and if statements

```python
def has_student_passed(score):
  """
  This is a multiline comment
  This function takes the score and returns pass if
  the score is above 49
  """
  if score>= 40:
    print("The student has passed the course.") # Gets executed if condition evaluates to true
  else:
    print("The student has not passed the course.") # Gets executed if condition evaluates to true
  return 0
```

For loop

```python
for s in scores:
  print(f"s={s}")
  has_student_passed(s)
```

Homework - Write a function to calculate grades and return the grade.

```python
def get_grade(score):
  """
  This is a multiline comment
  This function takes the score and returns the grade
  """
  if score>= 40 and score <= 50:
    return "E"
  elif score>50 and score<=60:
    return "D"
  elif score>60 and score<=100:
    return "A"
  return "U"
```

```python
for g in scores:
  print(f"Grade={g},{get_grade(g)}")
```
