Functions and if statements

```python
def has_student_passed(score):
  """
  This is a multiline comment
  This function takes the score and returns pass if
  the score is above 49
  """
  if score>= 40:
    print("The student has passed the course.") # Gets executed if condition evaluates to true
  else:
    print("The student has not passed the course.") # Gets executed if condition evaluates to true
  return 0
```

For loop

```python
for s in scores:
  print(f"s={s}")
  has_student_passed(s)
```

Homework - Write a function to calculate grades and return the grade.

```python
def get_grade(score):
  """
  This is a multiline comment
  This function takes the score and returns the grade
  """
  if score>= 40 and score <= 50:
    return "E"
  elif score>50 and score<=60:
    return "D"
  elif score>60 and score<=100:
    return "A"
  return "U"
```

```python
for g in scores:
  print(f"Grade={g},{get_grade(g)}")
```

Slicing arrays

```python
food_prices = np.array([530,293,200,501,293,706,877,291])
print(food_prices[1]) #prints the second element as indexing starts from zero.
print(food_prices[3:6]) #Prints elements from 4 to 6 (not 7). i.e. it doesn't include the 7th element.
print(food_prices[3:]) #Prints from the 4th element till the end of the array
print(food_prices[:3]) #From the start of the array to the 3rd elemnt. (whatever is after the colon is excluded)
print(food_prices[:-2]) #Print without the last two elements.
print(food_prices[-3:-1]) #Prints the last two elements.
print(food_prices[::2]) #Prints every second element.
print(food_prices[::3]) #Prints every third element.
print(food_prices[::-1]) #List is reversed
print(food_prices[::-2]) #List is reversed but it prints every second element
```
